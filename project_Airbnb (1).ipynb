### Variables Description 


**id:** A unique identifier for each listing.

**number_of_reviews:** The number of reviews that the listing has received.

**last_review:** The date of the last review that the listing received.

**reviews_per_month:** The average number of reviews that the listing receives each month.

**calculated_host_listings_count:** The total number of listings that the host has across all of their properties.

**availability_365:** The number of days per year that the listing is available for booking.
